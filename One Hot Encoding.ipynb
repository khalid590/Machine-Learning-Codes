```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
region = ['southwest', 'southeast', 'northwest', 'northeast']
```

```python
ordinal = OrdinalEncoder(categories = [region])
encoded = ordinal.fit_transform(df7[['region']])
```

```python
new_region = pd.DataFrame(encoded, columns = ['region'])  #Encode the value
```

```python
new_region.head()
```

```python
df7.region = new_region  #replace encoded value
```
